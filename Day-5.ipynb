### Robust Scaler
- If your dataset contains more no of outliers, then Robust Scaler is best techinque for scaling
- 

```python
# ("https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.RobustScaler.html")(about robust scaler)
```

```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
```

```python
data1.head()
```

```python
data1.columns
```

```python
data1.drop(['Unnamed: 0'], axis=1, inplace = True)
```

```python
data1.head()
```

### Robust Scaler Formula
 
 #####  x1 - Q1(x)
 ##### ------------------
 #####   Q3(x) - Q1(x)
  
 ### Inter Quartile Range (IQR)
 #####   Q1 ---->  25%(Lower Quartile)
 #####   Q2 ---->  50%(Medium Quartile)
 #####   Q3 ---->  75%(Upper Quartile)

## Apply Robust Scaler Model

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
rs
```

```python
rs2 = rs.fit_transform(data1)
rs2
```

```python
rs2 = pd.DataFrame(rs2, columns = data1.columns)
rs2.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2 , figsize = (9,5))
sns.kdeplot(data1['TV'], ax = ax1)
sns.kdeplot(data1['radio'], ax = ax1)
sns.kdeplot(data1['newspaper'], ax = ax1)
sns.kdeplot(data1['sales'], ax = ax1)

sns.kdeplot(rs2['TV'], ax = ax2)
sns.kdeplot(rs2['radio'], ax = ax2)
sns.kdeplot(rs2['newspaper'], ax = ax2)
sns.kdeplot(rs2['sales'], ax = ax2)
```

```python
data1.head()
```

```python
data1.quantile(0.25)
```

```python
data1.describe()
```

### Standard Scaler
- Most commonly used Scaling techinque. it scales between 0 to mean

```python
### Standard Scaler Formula 

#  xi - mean(x)
# ----------------
#   std(x)
```

```python
#  https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html (about standard scaler)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# step1: Import model
# step2: Initialize model into new object
# step3: fit the model to dataset
# step4: Convert array values into dataframe
```

```python
ss = StandardScaler()
ss
```

```python
ss2 = ss.fit_transform(data1)
ss2
```

```python
ss2 = pd.DataFrame(ss2, columns = data1.columns)
ss2.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2 , figsize = (9,5))
sns.kdeplot(data1['TV'], ax = ax1)
sns.kdeplot(data1['radio'], ax = ax1)
sns.kdeplot(data1['newspaper'], ax = ax1)
sns.kdeplot(data1['sales'], ax = ax1)

sns.kdeplot(ss2['TV'], ax = ax2)
sns.kdeplot(ss2['radio'], ax = ax2)
sns.kdeplot(ss2['newspaper'], ax = ax2)
sns.kdeplot(ss2['sales'], ax = ax2)
```
